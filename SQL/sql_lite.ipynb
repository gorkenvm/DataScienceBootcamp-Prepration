# LIKE ile Seçme İşlemleri

## LEFT JOIN & RIGHT JOIN

```python
#LEFT JOIN: Soldaki tablo ile kesişen sağdaki tablo değerlerinin bir araya getirilmesi. 
#Dikkat burası önemli. SQLite içerisndeki left join işlemleri LEFT OUTER JOIN ile yapılır.
#Bazı veri tabanlarınde direk LEFT JOIN ifadesi ile bazılarında ise LEFT OUTER JOIN ile yapılabilmektedir.
#Biz ikisinide kullandığımızda bir left join birleştirme işlemi yapılacaktır. 
#Bilginiz olsun.
```

```python
#öncelikle en kaba haliyle iki tabloyu nasıl birleştireceğimizi ele alalım:
```

```python
%sql select * from Products LEFT OUTER JOIN Categories ON Products.CategoryID = Categories.CategoryID LIMIT 5;
```

```python
#bu kadar arkadaşlar.
#evet bir araya aldık bunları ama bu sefer de sanki bir çıktı kalabaligi var.
#bunu nasil ortadan kaldiriiz.
#yani mesela ben sadece productname'e, categoryid'sine ve category ismine erişmek istersem ne yapıcam?
#hemen bunu ele alalım
```

```python
%sql select Products.ProductName, Categories.CategoryID, Categories.CategoryName from Products LEFT JOIN Categories ON Products.CategoryID = Categories.CategoryID LIMIT 5;
```

```python
#bakin tertemiz left join sonrasında sadece istediğimiz değişkenleri seçebildik. 
#görüyorum ve arttırıyorum. Bakın bu kod biraz çirkinleşmeye kalabalıklaşmaya başladı.
#daha fazla değişken seçmek ya da birden fazla peş peşe işlem gerçekleştirmek istediğimizde
#bu işlemler nasıl oalcak bunların daha temizi var mı dersek var
#şimdi aynı işlemi biraz daha temiz yazalım:
```

```python
%sql select PR.ProductName, CA.CategoryID, CA.CategoryName from Products as PR LEFT JOIN Categories as CA ON PR.CategoryID = CA.CategoryID LIMIT 5;
```

left join products p on od.product_id = p.product_id
left join categories ca on p.category_id = ca.category_id
left join customers cu on o.customer_id = cu.customer_id
where cu.country = 'Germany'
group by category_name;

left join order_details on order_details.product_id = products.product_id 
left join orders on order_details.order_id = orders.order_id 
WHERE order_date like '%07%' or order_date like '%08%' or order_date like '%09%';


#19 Müşterilerin 1997 yılına ait spiarişlerinin maksimum miktarlarını bulup müşterilerin isimleri ile birlikte buyukten kucuğe sıralayınız.
SELECT customer_name, MAX(price) maks FROM products 
left join order_details on products.product_id = order_details.product_id 
left join orders on order_details.order_id = orders.order_id
left join customers on customers.customer_id = orders.customer_id 
WHERE order_date like '%1997%' 
group by customer_name
order by maks DESC;

#20 1997 yılında çalışanları sipariş alma sayılarına göre sıralayınız

select count(*) n, em.first_name  from orders 
left join employees em on orders.employee_id = em.employee_id
WHERE order_date like '%1997%' 
group by em.first_name
ORDER by n DESC;
